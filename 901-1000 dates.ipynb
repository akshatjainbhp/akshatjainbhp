```python
df.apply(lambda x: x.astype(str).str.lower())
```

```python
import re
import datetime
aks=502
for ii in df['sentence'][900:1000]:
    me=ii
    add=0
    num=re.search(r'^((sunday|monday|tuesday|wednesday|thursday|friday|saturday)(\s)*)',ii)
    if num:
        b=num.group(2)
        b=b.title()
        d = datetime.date.today()
        
        while d.strftime('%A') != b:
            d+= datetime.timedelta(1)
            aa=str(d)
            
        san=re.search(r'(([0-9]{4})(-)([0-9]{1,2})(-)([0-9]{1,2}))',aa)
        y=san.group(2)
        m=san.group(4)
        da=san.group(6)
        date=da+"-"+m+"-"+y
        ii=date
        
    print(ii,"////////////",me)
    aks=aks+1
```
